```python
def create_table():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        cur=con.cursor()
        cur.execute("create table  if not exists st(admno int primary key, name varchar(30), dob date, pom decimal(4,2))")
        con.commit()
    except Exception as msg:
        print("Some Error",msg)
```

```python
def insert_rec():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        if con.is_connected():
            print("connected to MYSQL version",con.get_server_info())
            cur=con.cursor()
            ad=int(input("enter admno"))
            nm=input("enter name")
            db=input("enter date of birth")
            pm=float(input("enter pocket money"))
            q="insert into st values({0},'{1}','{2}',{3})".format(ad,nm,db,pm)
            cur.execute(q)
            con.commit()
    except Exception as msg:
        print("Some Error",msg)
    finally:
        if (con.is_connected()):
            cur.close()
            con.close()
            print("MySQL connection is closed")
```

```python
def search_rec():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        if con.is_connected():
            print("connected to MYSQL version",con.get_server_info())
            cur=con.cursor()
            ad=int(input("enter admno"))
            q="select * from st where admno={}".format(ad)
            cur.execute(q)
            for i in cur:
                print(i)
    except Exception as msg:
        print("Some Error",msg)
    finally:
        if (con.is_connected()):
            cur.close()
            con.close()
            print("MySQL connection is closed")
```

```python
#displaying all records using fetchall() method
def display_all_rec1():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        if con.is_connected():
            print("connected to MYSQL version",con.get_server_info())
            cur=con.cursor()
            q="select * from st"
            cur.execute(q)
            x=cur.fetchall()
            for i in x:
                print(i)
    except Exception as msg:
        print("Some Error",msg)
    finally:
        if (con.is_connected()):
            cur.close()
            con.close()
            print("MySQL connection is closed")
```

```python
#displaying all records using fetchone() method
def display_all_rec2():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        if con.is_connected():
            print("connected to MYSQL version",con.get_server_info())
            cur=con.cursor()
            q="select * from st"
            cur.execute(q)
            while True:
                x=cur.fetchone()
                if not x:break
                print(x)
    except Exception as msg:
        print("Some Error",msg)
    finally:
        if (con.is_connected()):
            cur.close()
            con.close()
            print("MySQL connection is closed")
```

```python
#displaying all records using fetchmany() method
def display_all_rec3():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        if con.is_connected():
            print("connected to MYSQL version",con.get_server_info())
            cur=con.cursor()
            q="select * from st"
            cur.execute(q)
            while True:
                x=cur.fetchmany(2)
                if not x:break
                print(x)
    except Exception as msg:
        print("Some Error",msg)
    finally:
        if (con.is_connected()):
            cur.close()
            con.close()
            print("MySQL connection is closed")
```

```python
def delete_rec():
    try:
        con=mysql.connector.connect(user="root",password="123",host="localhost",database="test")
        if con.is_connected():
            print("connected to MYSQL version",con.get_server_info())
            cur=con.cursor()
            nm=input("enter name")
            q="delete from st where name='{}'".format(nm)
            cur.execute(q)
            con.commit()
    except Exception as msg:
        print("Some Error",msg)
    finally:
        if (con.is_connected()):
            cur.close()
            con.close()
            print("MySQL connection is closed")
```
